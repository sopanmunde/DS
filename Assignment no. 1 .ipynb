### Filling missing values Director and Cast with NoDataAvailable.

```python
df[['director', 'cast']]=df[['director', 'cast']].fillna('NoDataAvailab')
```

### Filling missing values 'country' and 'date added' with mode()â€™

```python
df['country'].fillna(df['country'].mode()[0], inplace=True)
df['date_added'].fillna(df['date_added'].mode()[0], inplace=True)
df
```

### In which year the highest number of the TV Shows and Movies were released? Show with bar graph.

```python
# Number of TV Shows and Movies released each year
release_counts = df['release_year'].value_counts().sort_index()

plt.figure(figsize=(12,6))
plt.bar(release_counts.index, release_counts.values, color='green')
plt.xlabel('Release Year')
plt.ylabel('Number of Releases')
plt.title('Number of TV Shows and Movies Released Per Year')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

max_year = release_counts.idxmax()
max_count = release_counts.max()
plt.axvline(max_year, color='red', linestyle='--', label=f'Peak Year: {max_year} ({max_count} releases)')
plt.legend()
plt.show()
max_year, max_count
```

###  Show Top 10 Directors,who give the highest no of TV Shows and Movies on Netflix?

```python


# Count the number of shows/movies by each director and get the top 10
directors = df_cleaned['director'].value_counts().head(10)

# Create the bar plot using seaborn
plt.figure(figsize=(12, 6))
sns.barplot(x=directors.index, y=directors.values, palette="Reds_r")

# Customize the plot
plt.xlabel('Director')
plt.ylabel('Number of Movies & TV Shows')
plt.title('Top 10 Directors with the Most Content on Netflix')
plt.xticks(rotation=45, ha='right')

# Show the plot
plt.show()
```

# Part 3: Data Visualisation

```python
# Filter dataset for titles containing Damsel
damsel_counts = df_cleaned[df_cleaned['title'].str.contains("Damsel", case=False, na=False)]['type'].value_counts()

plt.figure(figsize=(6, 4))
sns.barplot(x=damsel_counts.index, y=damsel_counts.values, palette="Blues_r")

plt.xlabel('Type')
plt.ylabel('Count')
plt.title('Number of Movies and TV Shows with Title "Damsel"')
plt.xticks(rotation=0)
plt.show()

damsel_counts
```

```python
# Count the number of TV Shows and Movies by each country
country_counts = df_cleaned['country'].value_counts().head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=country_counts.index, y=country_counts.values, palette="viridis")

plt.xlabel('Country')
plt.ylabel('Number of TV Shows & Movies')
plt.title('Top 10 Countries with the Highest Number of TV Shows & Movies on Netflix')
plt.xticks(rotation=45, ha='right')

plt.show()            # Show the plot

country_counts      # Display the top 10 countries with counts
```

### Top 10 countries with the highest number of movies.

```python
# Count the number of Movies by each country
movie_counts_by_country = df_cleaned[df_cleaned['type'] == 'Movie']['country'].value_counts().head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=movie_counts_by_country.index, y=movie_counts_by_country.values, palette="magma")

plt.xlabel('Country')
plt.ylabel('Number of Movies')
plt.title('Top 10 Countries with the Highest Number of Movies on Netflix')
plt.xticks(rotation=45, ha='right')

plt.show()

movie_counts_by_country
```

### Growth of Netflix Content in Different Countries â€“ Use a line graph to compare how many movies/TV shows have been added over the years for different countries.


```python
# Select top 5 countries with the most content
top_countries = df_cleaned['country'].value_counts().head(5).index


country_year_data = df_cleaned[df_cleaned['country'].isin(top_countries)]       # Filter data for selected countries

country_trend = country_year_data.groupby(['release_year', 'country']).size().unstack()    # Count the number of Movies & TV Shows added over the years for each country

plt.figure(figsize=(12, 6))            # Plot the line graph using Seaborn
sns.lineplot(data=country_trend, marker='o')

plt.xlabel('Year')
plt.ylabel('Number of Movies & TV Shows')
plt.title('Growth of Netflix Content in Different Countries Over the Years')
plt.legend(title="Country", loc="upper left")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
